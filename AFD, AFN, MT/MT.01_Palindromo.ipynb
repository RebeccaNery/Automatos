```python
#AUTÔMATO QUE RECONHECE SE A STRING BINÁRIA É UM PALÍNDROMO

def is_palindrome(binary_string):
  return binary_string == binary_string[::-1]

binary_string = '101'

if is_palindrome(binary_string):
  print(f"A sequência {binary_string} é um palíndromo!")
else:
  print(f"A sequência {binary_string} >>> NÃO <<< é um palíndromo!")

#Explicação:
# 1. Usamos [::-1] para inverter a string
# 2. Comparamos a versão original da string com sua versão invertida
# 3. Se as duas versões forem iguais, é um palíndromo
```
