```python
# Máquinas de Turing - Exemplo 02.
# Exemplo 2: Multiplicação Unária por Repetição

def unary_multiplication(a, b):
    # Representa 'a' e 'b' em unário como strings de '1's
    unary_a = "1" * a
    unary_b = "1" * b

    # Multiplica ao concatenar 'b' para cada '1' em 'a'
    result = unary_b * a

    # Retorna o resultado e também a representação em decimal
    return result, len(result)

# Testando a função
a, b = 3, 2  # Representa 3 * 2
result_unary, result_decimal = unary_multiplication(a, b)
print(f"Unário: {result_unary} (representa {result_decimal} em decimal)")

# Explicação:
# 1.	unary_a e unary_b são representações unárias de aaa e bbb.
# 2.	Multiplicamos ao concatenar unary_b a a vezes, simulando a adição repetida.
# 3.	O comprimento da string resultante em unário representa o produto a×ba \times ba×b em decimal.

# Comparação e Eficiência
# Essas implementações são muito mais rápidas e eficientes do que uma simulação completa de
# uma Máquina de Turing, pois o Python usa operações diretas de strings e manipulação de listas.
# Esses códigos são adequados para cenários onde a simplicidade e a eficiência são
# preferidas sobre a fidelidade ao modelo teórico da Máquina de Turing.
```
