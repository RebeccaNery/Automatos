```python
#AUTÔMATO PARA MASCARAMENTO DE CPF
def mascara_cpf (cpf):
  estado = 'q0'

  for char in cpf:
    if estado == 'q0':
      char = 'X'
      estado = 'q1'
    elif estado == 'q1':
      char = 'X'
      estado = 'q2'
    elif estado == 'q2':
      char = 'X'
      estado = 'q3'
    elif estado == 'q3':
      estado = 'q4'
    elif estado == 'q4':
      estado = 'q5'
    elif estado == 'q5':
      estado = 'q6'
    elif estado == 'q6':
      estado = 'q7'
    elif estado == 'q7':
      estado = 'q8'
    elif estado == 'q8':
      estado = 'q9'
    elif estado == 'q9':
      char = 'X'
      estado = 'q10'
    elif estado == 'q10':
      char = 'X'
      estado = 'q_aceita'
    elif estado == 'q_aceita':
      return "CPF mascarado com sucesso! ==>"
    else:
      estado = 'q_rejeita'
      return "CPF inválido, tente novamente."
      break
    print(char, end='')
```

```python
#Testes

print(mascara_cpf('99999999999'))
```
