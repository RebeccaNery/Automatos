```python
#AUTÃ”MATO PARA RECONHECIMENTO DE PLACAS DE CARRO ESTILO MERCOSUL

# 3 letras / 1 nÃºmero / 1 letra / 2 nÃºmeros

# funÃ§Ã£o para verificar se o caracter Ã© uma letra ==> nome_string.isalpha()

def reconhecer_placa_mercosul(digitos):
  estado = 'q0'

  for char in digitos:
    if estado == 'q0':
      if digitos[0].isalpha():
        estado = 'q1'
      else:
        estado = 'q_rejeita'
    elif estado == 'q1':
      if digitos[1].isalpha():
        estado = 'q2'
      else:
        estado = 'q_rejeita'
    elif estado == 'q2':
      if digitos[2].isalpha():
        estado = 'q3'
      else:
        estado = 'q_rejeita'
    elif estado == 'q3':
      if digitos[3].isdigit():
        estado = 'q4'
      else:
        estado = 'q_rejeita'
    elif estado == 'q4':
      if digitos[4].isalpha():
        estado = 'q5'
      else:
        estado = 'q_rejeita'
    elif estado == 'q5':
      if digitos[5].isdigit():
        estado = 'q6'
      else:
        estado = 'q_rejeita'
    elif estado == 'q6':
      if digitos[6].isdigit():
          estado = 'q_aceita'
      else:
        estado = 'q_rejeita'
    print(char, end='')


  if estado == 'q_aceita':
    return "A PLACA > " + digitos + " < ESTÃ EM CONFORMIDADE COM OS PARÃ‚METROS DO MERCOSUL."


  if estado == 'q_rejeita':
    return "A PLACA > " + digitos + " < NÃƒO CORRESPONDE AOS PARÃ‚METROS DO MERCOSUL."


```
